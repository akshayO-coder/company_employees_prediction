# #Data cleaning, preprocessing, and splitting into training and testing

```python
df.head(5)
```

```python
df.columns
```

```python
X_cat = df[['BusinessTravel','Department','EducationField','Gender','MaritalStatus','JobRole']]
```

```python
x_cat = df[['BusinessTravel','Department','EducationField','Gender','MaritalStatus','JobRole']]
```

```python
x_cat
```

```python
from sklearn.preprocessing import OneHotEncoder
onehotencoder = OneHotEncoder()
x_cat = onehotencoder.fit_transform(x_cat).toarray()
```

```python
x_cat
```

```python
x_cat.shape
```

```python
x_cat = pd.DataFrame(x_cat)
```

```python
x_cat
```

```python
x_numerical = df[['Age', 'DailyRate', 'DistanceFromHome','Education', 'EnvironmentSatisfaction', 'HourlyRate', 'JobInvolvement','JobLevel','JobSatisfaction','MonthlyIncome','MonthlyRate','NumCompaniesWorked','OverTime','PercentSalaryHike', 'PerformanceRating','RelationshipSatisfaction','StockOptionLevel','TotalWorkingYears','TrainingTimesLastYear', 'WorkLifeBalance','YearsAtCompany','YearsInCurrentRole', 'YearsSinceLastPromotion','YearsWithCurrManager']]
```

```python
X_all = pd.concat([x_cat, x_numerical], axis=1)
```

```python
X_all
```

# #scaling the dataset for machine learning

```python
from sklearn.preprocessing import MinMaxScaler
scaler = MinMaxScaler()
X = scaler.fit_transform(X_all)
```

```python
X
```

```python
y = df['Attrition']
```

```python
y
```

```python
# splitting the dataset

from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size = 0.30)
```

```python
X_train.shape
```

```python
X_test.shape
```

# LogisticRegression

```python
# performing machine learning models
from sklearn.linear_model import LogisticRegression
from sklearn.metrics import accuracy_score

```

```python
model = LogisticRegression()
model.fit(X_train, y_train)
```

```python
y_pred = model.predict(X_test)
```

```python
y_pred
```

```python
from sklearn.metrics import confusion_matrix, classification_report

print(f'accuracy is {100*accuracy_score(y_pred, y_test)}%')
```

```python
cm = confusion_matrix(y_pred, y_test)
sns.heatmap(cm, annot= True)
legend = False
```

```python
print(classification_report(y_test, y_pred))
```

# RandomForestClassifier

```python
from sklearn.ensemble import RandomForestClassifier
model = RandomForestClassifier()
model.fit(X_train, y_train)
```

```python
X_train.shape
```

```python
y_train.shape
```

```python
y_pred = model.predict(X_test)
```

```python
y_pred
```

```python
cm = confusion_matrix(y_pred, y_test)
sns.heatmap(cm, annot = True)
legend = False
```

```python
print(classification_report(y_test, y_pred))
```
